Las Variables orbital_period, Mass y Distance presentan Valores Ausentes o Atipicos

```python
#para la variable orbital_period puedo completar con la mediana ya que presenta pocos valores (sin informacion)
df["orbital_period"].fillna(df["orbital_period"].median(), inplace = True)

#para la variable distance puedo usar la moda para promediar y completar los valores (sin informacion)
most_frequent_distance = df["distance"].mode()[0]
df["distance"].fillna(most_frequent_distance, inplace=True)

#para la variable mass utilizo nuevamente la moda para promediar y completar los valores (sin informacion)
most_frequent_mass = df["mass"].mode()[0]
df["mass"].fillna(most_frequent_mass, inplace=True)


df.head()
```

Una vez visualizado se genera una funcion para la limpieza de outliers.
Usamosa la funcion en las columnas numericas. Luego los concatenamos y los eliminamos delconjunto de Datos.
Imprimimos en pantalla la cantidad de outliers detectados.

```python
def detect_outliers_iqr(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 0.5 * iqr
    upper_bound = q3 + 0.5 * iqr
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

columns_to_check = ["orbital_period"]
outliers_list = []
for col in columns_to_check:
    outliers = detect_outliers_iqr(df, col)
    outliers_list.append(outliers)

outliers_df = pd.concat(outliers_list)

df_cleaned1 = df.drop(outliers_df.index)

print("Número de outliers detectados:", len(outliers_df))
print(outliers)

```
